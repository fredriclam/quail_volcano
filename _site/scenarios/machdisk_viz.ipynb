```python
import numpy as np
import matplotlib.colors
import matplotlib.pyplot as plt
import os
```

```python
data = np.load("temp_data.npz")
list(data.keys())
```

```python
Z = data["dat"]
t = data["t"]

# Distribute packed array to named arrays
y, Mx, My, p, rho, c, T = np.moveaxis(Z, 2, 0)
# Flatten y
y = y[0,:]
c.shape, t.shape, y.shape
```

```python
# Shifted diverging cmap for Mach
shift_div_cmap = matplotlib.colors.LinearSegmentedColormap.from_list(
    f'trunc(bwr,{0.25},{1.0})',
    matplotlib.cm.bwr(np.linspace(0.25, 1, 100)))
```

```python
mg_y, mg_t = np.meshgrid(y, t)
fig, ax = plt.subplots(figsize=(8,6))
plt.contourf(mg_t, mg_y, My, levels=np.linspace(0.0, 3.0, 100), cmap=shift_div_cmap)
ax.set_xlabel("$t$ (s)")
ax.set_ylabel("$z$ (m)")
cb = plt.colorbar()
cb.set_label("M_z")
```

```python
mg_y, mg_t = np.meshgrid(y, t)
fig, ax = plt.subplots(figsize=(8,6))
plt.contourf(mg_t, mg_y, p, levels=10)
ax.set_xlabel("$t$ (s)")
ax.set_ylabel("$z$ (m)")
cb = plt.colorbar()
cb.set_label("M_z")
```
